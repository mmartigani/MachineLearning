```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
dataset
```

```python
#lo que vamos hacer ahora atraves de un bucle for es 
#eliminar del dataset  los valores nan
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i, j]) for j in range(0,20)])#20 son las columnas 
```

```python
rules = apriori(transactions, min_support=0.003, min_confidence=0.2)
```
